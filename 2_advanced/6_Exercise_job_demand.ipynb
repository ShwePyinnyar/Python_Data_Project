```python
import pandas as pd
from datasets import load_dataset
import matplotlib.pyplot as plt

dataset=load_dataset('lukebarousse/data_jobs')
df=dataset['train'].to_pandas()

df.job_posted_date=pd.to_datetime(df.job_posted_date)
```

```python
df_us = df[df.job_country=='United States'].copy()
df_us.head()
```

```python
df_us['Month_no'] = df_us['job_posted_date'].dt.month
df_us.head()
```

```python
df_us['Month_name'] = df_us['job_posted_date'].dt.month_name()
df_us.head()
```

```python
jobs = ['Data Analyst','Data Scientist', 'Data Engineer']
df_us = df_us[df_us['job_title_short'].isin(jobs)]

df_us

```

```python
pivot_month_size = df_us.pivot_table(index='Month_no',columns='job_title_short',aggfunc='size')
pivot_month_size.index = pd.to_datetime(pivot_month_size.index, format='%m').month_name()
```

```python
pivot_month_size
```

```python
a = pivot_month_size.plot()
a.set_xticks(range(len(pivot_month_size.index)))
a.set_xticklabels(pivot_month_size.index)
plt.xlabel('2023')
plt.ylabel('Number counts')
plt.title('Job count trend lines')
plt.xticks(rotation=45, ha='right')
plt.legend()
plt.show()
```
