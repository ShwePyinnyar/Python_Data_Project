```python
import pandas as pd
from datasets import load_dataset
import matplotlib.pyplot as plt

dataset=load_dataset('lukebarousse/data_jobs')
df=dataset['train'].to_pandas()

df.job_posted_date=pd.to_datetime(df.job_posted_date)
```

```python
df.pivot_table(values='salary_year_avg',index='job_country',columns='job_title_short',aggfunc='median')
```

```python
top_countries = df.job_country.value_counts().head(6).index
top_countries
```

```python
jobs = ['Data Analyst','Data Engineer','Data Scientist']
```

```python
df_job_country = df.pivot_table(values='salary_year_avg',index='job_country',columns='job_title_short',aggfunc='median')
df_job_country = df_job_country.loc[top_countries]
df_job_country = df_job_country[jobs]

df_job_country
```

```python
df_job_country.plot(kind='bar')
plt.ylabel('Median Salary')
plt.xlabel('')
plt.title('Median Salary per Country')
plt.xticks(rotation=45,ha='right')
plt.show()
```
